```python
# to dictionary poy antoisthxei sto aggliko alfavito
ab = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4,
    'e': 5,
    'f': 6,
    'g': 7,
    'h': 8,
    'i': 9,
    'j': 10,
    'k': 11,
    'l': 12,
    'm': 13,
    'n': 14,
    'o': 15,
    'p': 16,
    'q': 17,
    'r': 18,
    's': 19,
    't': 20,
    'u': 21,
    'v': 22,
    'w': 23,
    'x': 24,
    'y': 25,
    'z': 26,
    ' ': 27,
    '.': 28,
    ',': 29,
    '?': 30,
    '!': 31,
    ':': 32

}

def unique_ch(phrase):
    # edw tha kratame poses fores emfanizetai ena gramma, sthn antistoixh thesh tou pinaka
    # arxikopoiw ton pinaka epishs
    c_list = [0] * 32

    # apla ektipwnw to kathe to key pou antistoixei se kathe gramma
    for letter in phrase:
        # auksanw kata ena thn emfanhsh tou grammatos sthn thesi tis listas pou antistoixei sto gramma
        c_list[ab.get(letter.lower())-1] += 1
        
    # counter gia na metrhsw tous monadikous xarakthres
    unique_counter = 0
    
    # psaxnw mesa sthn lista me tis emfanhseis twn xarakthrwn, posoi xarakthres emfanizontai toulaxiston mia fora
    for index in range(1,32):
        if c_list[index] != 0:
            unique_counter += 1
            
    # epistrefw to sunolo twn monadikwn xarakthrwn
    return print("uparxoun " + str(unique_counter) + " monadikoi xarakthres");

    

print(unique_ch(input("Enter your message: ")))
        
        
    
    
    
    
    
    
```
